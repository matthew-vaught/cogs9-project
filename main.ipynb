```python
res = requests.get('https://sandman.sleepcycle.com/data/per-country-stats')
json = res.json()
```

```python
sleep_df = pd.DataFrame(json)
sleep_df.head()
```

```python
sleep_df['country'] = sleep_df['country'].str.upper()
abb_path = Path('data') / 'country_abbrev_list.csv'
country_abbrevs = pd.read_csv(abb_path)
sleep_df = pd.merge(sleep_df, country_abbrevs, left_on = 'country', right_on = 'Code').drop(['country', 'Code'], axis=1).set_index('Name')
```
